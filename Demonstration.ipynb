### Example 5

Convolve model with sinus signal

```python
signal = np.sin(np.linspace(-1.5, 4.7, 100))

d_func = np.zeros(1000)
for i in [43, 75, 157, 354, 400, 754, 758]:
    d_func[i] = np.random.uniform(1, 3)
    
test5 = np.convolve(d_func, signal, mode='same')
test5 = test5 + 5 + np.random.uniform(0, 1, len(test5))

fig = plt.figure(figsize=(7, 5))
plt.plot(test5)
plt.xlabel('$Counts [Ed]$')
plt.ylabel('$Amplitude [Ed]$')
plt.grid()
plt.show()
```

```python
step = 50
points5 = get_borders(test5, step)
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(test5)
plt.grid()
plt.xlabel('$Counts [Ed]$')
plt.ylabel('$Amplitude [Ed]$')
for i in points5:
    plt.plot([i, i], [1, 11], c='r')
plt.show()
```

# Real data

### Example 1

```python
df = pd.read_csv('data/130.csv', sep=',', 
                 index_col=False, skiprows=[1, 2])
df = df.replace([-999.25], 0)
```

```python
fig, ax = plt.subplots(figsize=(5, 7))
plt.plot(df['RSN'], df['MD'], alpha=0.8)
plt.grid()
plt.xlabel('$Amplitude [Ed]$')
plt.ylabel('$Measured Depth [m]$')
ax.invert_yaxis()
fig.show()
```

```python
step = 1000  # шаг по массиву
diff = 10  # процент отклонения
filter_value = 1000

points11 = get_borders(df['RSN'], step, diff, filter_value)
```

```python
fig, ax = plt.subplots(figsize=(5, 7))
plt.plot(df['RSN'], df['MD'], alpha=0.8)
for i in points11:
    plt.plot([0, 100], [df['MD'][i], df['MD'][i]], c='r')
plt.grid()
plt.xlabel('$Amplitude [Ed]$')
plt.ylabel('$Measured Depth [m]$')
ax.invert_yaxis()
plt.show()
```

### Example 2

```python
from datetime import datetime


df = pd.read_csv('data/S&P 500 Historical Data00-20.csv', sep=',')
data_oy = []
data_ox = []

for i in range(len(df['Low'])):
    data_oy.append(float(df['Low'][i].replace(',', '')))
    data_ox.append(datetime.strptime(df['Date'][i], '%m/%d/20%y'))
    
data_oy = data_oy[::-1]
data_ox = data_ox[::-1]

fig = plt.figure(figsize=(7, 5))
plt.plot(data_ox, data_oy, alpha=0.8)
plt.xlabel('$Date [Year]$')
plt.ylabel('$Price [$]$')
plt.grid()
fig.show()
```

```python
step = 100  # шаг по массиву
diff = 20

points22 = get_borders(data_oy, step, diff)
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(data_ox, data_oy, alpha=0.8)
for i in points22:
    plt.plot([data_ox[i], data_ox[i]], [100, 3000], c='r')
plt.xlabel('$Date [Year]$')
plt.ylabel('$Price [$]$')
plt.grid()
fig.show()
```
