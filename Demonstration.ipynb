```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from algorithm import get_borders
```

## Example 1 (synthetics)

```python
test1 = [10 for i in range(30)] + [14 for i in range(30)] + [10 for i in range(30)]
test1 = np.array(test1) + np.random.sample(len(test1))

fig = plt.figure(figsize=(7, 5))
plt.plot(test1)
plt.grid()
plt.show()
```

```python
step = 3  # шаг по массиву
diff = 5  # процент отклонения 
points1 = get_borders(test1, step, diff)
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(test1)
plt.grid()
for i in points1:
    plt.plot([i, i], [10, 15], c='r')
plt.show()
```

## Example 4 (real data)

```python
df = pd.read_csv('data/130.csv', sep=',', 
                 index_col=False, skiprows=[1, 2])
df = df.replace([-999.25], 0)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(df['MD'], df['RSN'], alpha=0.8)
plt.grid()
fig.show()
```

```python
step = 1000  # шаг по массиву
diff = 10  # процент отклонения
filter_value = 1000

points4 = get_borders(df['RSN'], step, diff, filter_value)
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(df['MD'], df['RSN'], alpha=0.8)
for i in points4:
    plt.plot([df['MD'][i], df['MD'][i]], [0, 100], c='r')
plt.grid()    
plt.show()
```
