```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mean(array):
    return sum(array)/len(array)


def running_mean(x, N):
    out = np.zeros_like(x, dtype=np.float64)
    for i in range(len(x)):
        if N%2 == 0:
            a, b = i - (N-1)//2, i + (N-1)//2 + 2
        else:
            a, b = i - (N-1)//2, i + (N-1)//2 + 1
        a = max(0, a)
        b = min(len(x), b)
        out[i] = np.mean(x[a:b])
    return list(out)


def func1(data, step, diff):
    all_mean = abs(mean(data))
    
    points = []
    interval = list(data[:step])
    for i in range((len(data) // step) - 1):

        current_interval = list(data[(i+1)*step:(i+2)*step])
        if (abs(mean(current_interval) - mean(interval))/all_mean)*100 < diff:
            interval += current_interval
        else:
            points.append((i+1)*step)
            interval = current_interval
            
    return points
```

## Example 1

```python
test1 = [10 for i in range(30)] + [14 for i in range(30)] + [10 for i in range(30)]
test1 = np.array(test1) + np.random.sample(len(test1))

fig = plt.figure(figsize=(7, 5))
plt.plot(test1)
plt.grid()
plt.show()
```

```python
step = 3  # шаг по массиву
diff = 5  # процент отклонения 
points1 = func1(test1, step, diff)
```

```python
fig = plt.figure(figsize=(7, 5))
plt.plot(test1)
plt.grid()
for i in points1:
    plt.plot([i, i], [10, 15], c='r')
plt.show()
```
